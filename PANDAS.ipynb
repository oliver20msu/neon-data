```python
# Make a dataframe out of a dictionary

import pandas as pd
adamprj={'codenames':['Antares','Megalon','Dark Side','Flip Side','Bodyguard','Agent'],
         'origin':[338,'1929B','Moon of Angoll','X sector','1661','72C'],
         'index':[8,7.2,6.0,4.9,7.2,8.0]}
adam_df=pd.DataFrame(adamprj)
print(adam_df.head())
```

```python
# Create a new dataframe out of just one column

slice=adam_df[['origin']]
print(slice.head())
```

```python
# Create new dataframe out of multiple columns

slices=adam_df[['codenames','origin']]
print(slices.head())
```

```python
# Rearrange columns

reorder=adam_df[['codenames','index','origin']]
print(reorder.head())
```
